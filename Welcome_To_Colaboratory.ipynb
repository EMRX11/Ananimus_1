```

num_col=df.select_dtypes(include=['float','int']).columns
df1=pd.pivot_table(df, index=['Country' , 'City'], values=num_col,aggfunc='mean')
df1
```

```
df1.loc[['Germany','Uzbekistan','Zimbabwe']]
```

```
df2=df1.copy()
df2
```

```
df2['AirPollution']=100-df2['AirQuality']
df2
```

```
df2['WaterQuality']=100-df2['WaterPollution']
df2
```

```
df3=pd.DataFrame(df2.values , index=df2.index , columns=[['Air','Air','Water','Water'],['AirQuality','AirPollution','WaterQuality','WaterPollution']])
df3
```

```
df3.columns.names=['Object','Emrx_11_08']
df3
```

```
# @title Air Pollution vs Water Quality Correlation

import matplotlib.pyplot as plt
plt.scatter(df3['Air']['AirPollution'], df3['Water']['WaterQuality'])
plt.xlabel('Air Pollution')
_ = plt.ylabel('Water Quality')
```
